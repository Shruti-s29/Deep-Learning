In sklearn for neural network implementation we use MLP classifier that is Multi-Layer Perceptron.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
# Loading the data
iris_data = datasets.load_iris()
iris_data
```

```python
x = iris_data.data
y = iris_data.target

# spliting 
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=1)
```

```python
# importing neural network classifier
from sklearn.neural_network import MLPClassifier
```

```python
# making classifier object 
clf = MLPClassifier()

# training
clf.fit(x_train,y_train)
```

```python
# finding the training score
clf.score(x_test,y_test)
```

```python
# Basically all the weights, it does not include biases
clf.coefs_
```

```python
# for biases
clf.intercepts_
```
